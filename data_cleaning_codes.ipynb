```python
def format_price(x):
    if x.endswith('/year'):
        return int(x.replace('/year', '').replace(',', ''))
    elif x.endswith('/month'):
        return int(x.replace('/month', '').replace(',', '')) * 12
    else:
        return int(x.replace(',', ''))
```

```python
df.Price = df.Price.apply(format_price)
```
